```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
churn = pd.read_csv(r"C:\Users\Tayna Appel\Documents\GitHub\Aulas - Ironhack\Labs\lab-handling-imbalance\files_for_lab\Customer-Churn.csv")
churn.head()
```

```python
churn.dtypes
```

```python
churn['TotalCharges'] = pd.to_numeric(churn['TotalCharges'],errors='coerce')
null_counts = churn.isnull().sum()
null_counts
```

```python
churn_clean = churn.dropna(subset=['TotalCharges'])
```

```python
features = ['tenure','SeniorCitizen','MonthlyCharges','TotalCharges']
X =churn_clean[features]
y = churn_clean['Churn'].map({'No': 0, 'Yes': 1})
```

```python
churn_clean
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

accuracy
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import imblearn
```

```python
from sklearn.utils import resample
```

```python
#imbalance only in the training data, never testing
```

```python
X_train_df = pd.DataFrame(X_train, columns=features)
training_data = pd.concat([X_train_df, y_train.reset_index(drop=True)], axis=1)
training_data.head()
```

```python
majority = training_data[training_data['Churn'] == 0]
minority = training_data[training_data['Churn'] == 1]
```

```python
majority_downsampled = resample(majority,
                                replace=False,
                                n_samples=len(minority),
                                random_state=0)
```

```python
df_downsampled = pd.concat([majority_downsampled, minority])
```

```python
X_down = df_downsampled[features]
y_down = df_downsampled['Churn']
```

```python
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(X_down, y_down, test_size=0.2, random_state=42, stratify=y_down)
```

```python
model_d = LogisticRegression()
model_d.fit(X_train_d, y_train_d)
```

```python
y_pred_d = model_d.predict(X_test_d)
accuracy_d = accuracy_score(y_test_d, y_pred_d)

accuracy_d
```
